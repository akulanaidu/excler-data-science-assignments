```python
import pandas as pd

# loading data set 
df = pd.read_csv('sales_data_with_discounts.csv')
df
```

```python
df.info()
```

```python
# Identify numerical columns in the dataset.

numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
print(numerical_columns)
```

```python

# Calculate the mean, median, mode, and standard deviation for the numerical columns
df_numerical = df[numerical_columns]

# Mean
mean = df_numerical.mean()

# Median
median = df_numerical.median()

# Mode
mode = df_numerical.mode().iloc[0]

# Standard deviation
std = df_numerical.std()

# Print the results
print("Mean:")
print(mean)
print("\nMedian:")
print(median)
print("\nMode:")
print(mode)
print("\nStandard Deviation:")
print(std)
```

```python

# By interpreting these statistics, we can gain insights into the mean,median ,mode,Standard deviatian of the numerical data in the sales dataset. This information can be helpful for understanding patterns and trends in the data, as well as for making predictions and informed decisions.
```

```python
#Plotting histograms for each numerical column.

import matplotlib.pyplot as plt

# Plot histograms for each numerical column
for column in numerical_columns:
    plt.hist(df[column])
    plt.title(f"Histogram of {column}")
    plt.show()
```

```python
#skewness :- Volume is positive, Avg Price is  positive, Total Sales Value is positive, Discount Rate (%) is negative, Discount Amount is positive, Net Sales Value is positive.
```

```python
# Create boxplots for numerical variables

for column in numerical_columns:
    plt.boxplot(df[column])
    plt.title(f"Boxplot of {column}")
    plt.show()
```

```python
categorical_columns = df.select_dtypes(include=['object']).columns
print ('categorical cloumns',[categorical_columns])
```

```python
# BarChart Analysis for Categorical Column and count of each category for above data

# Count the number of occurrences for each category in the categorical columns
for column in categorical_columns:
  unique_categories = df[column].unique()
  category_counts = df[column].value_counts()

  # Create a bar chart for each categorical column
  plt.bar(unique_categories, category_counts)
  plt.title(f"Bar Chart for {column}")
  plt.xlabel(column)
  plt.ylabel("Count")
  plt.show()

  # Print the count of each category
  print(f"Count of categories in {column}:")
  print(category_counts)
```

```python
#Standardization (z-score normalization) is a technique used to transform numerical variables to have a mean of 0 and a standard deviation of 1. This makes the variables more comparable and helps to improve the performance of many analytical models."
```

```python
# Standardize the data
for column in numerical_columns:
  mean = df[column].mean()
  std = df[column].std()
  zscore_df=(df[column] - mean) / std


# Print the standardized data
df[numerical_columns].head()
```

```python
# Explain the need for converting categorical data into dummy variables
# Apply one-hot encoding to the categorical columns
df_encoded = pd.get_dummies(df, columns=categorical_columns)

# Display a portion of the transformed data
print(df_encoded.head())
```
